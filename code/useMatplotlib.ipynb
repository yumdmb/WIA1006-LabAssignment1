```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../dataset/House Pricing.csv')

# sccatter plot for price vs. area
plt.scatter(df['area'], df['price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Price vs. Area')
plt.show()

# scatter plot for price vs. bedrooms
plt.scatter(df['bedrooms'], df['price'])
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.title('Price vs. Bedrooms')
plt.show()

```
